# Books Sales - EDA

---
The objective of this project is to perform Exploratory Data Analysis (EDA) on the book sales dataset to understand trends in sales, ratings, and revenue, analyze patterns across authors, genres, and publishers, detect anomalies, and generate insights for better decision-making.

---
### Importing Libararies

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

---
### Loading Dataset

```python
df= pd.read_csv("Books_Data_Clean.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df= df[df["Publishing Year"]>1900]
```

```python
df.isna().sum()
```

```python
df.dropna(subset="Book Name", inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.nunique()
```

---
**What does the distribution of publishing years tell us about book releases over time?**

```python
plt.hist(df["Publishing Year"])
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("Distribution of Publishing Year")
```

---
**Which genre has the highest number of books published?**

```python
df["genre"].value_counts().plot(kind="bar")
plt.xlabel("Genre")
plt.ylabel("Number of Books")
plt.title("Number of Books in Each Genre")
```

---
**Which authors have the highest average book ratings?**

```python
df.groupby("Author")["Book_average_rating"].mean().sort_values(ascending=False)
```

---
**How do book rating counts vary across different genres?**

```python
sns.boxplot(x="genre", y="Book_ratings_count", data= df)
plt.xlabel("Genre")
plt.ylabel("Book Rating Count")
plt.title("Box Plot for Book ratings Count for each genre")
```

---
**What is the relationship between sale price and units sold?**

```python
plt.scatter(df["sale price"], df["units sold"])
plt.xlabel("Sale Price")
plt.ylabel("Units Sold")
plt.title("Scatter Plot of Sale Price vs Units Sold")
```

---
**What is the distribution of books across different languages?**

```python
language_counts= df["language_code"].value_counts()
plt.pie(language_counts, labels=language_counts.index, startangle=90, autopct="%1.1f%%")
plt.title("Language Distribution of Books")
```

---
**Which publishers generate the highest total revenue?**

```python
df.groupby("Publisher ")["publisher revenue"].sum().sort_values(ascending=False)
```

---
**What is the maximum average book rating count across different author ratings?**

```python
df.groupby("Author_Rating")["Book_ratings_count"].mean().sort_values(ascending=False).max()
```

---
**Which language has the highest number of books published?**

```python
df.groupby("language_code").size().sort_values(ascending=False)
```

---
**What is the maximum number of book ratings received for each author rating?**

```python
df.groupby("Author_Rating")["Book_ratings_count"].max()
```

---
**How does a book’s average rating relate to the number of ratings it receives?**

```python
plt.scatter(df["Book_average_rating"], df["Book_ratings_count"])
plt.xlabel("Book Average Rating")
plt.ylabel("Book Ratings Count")
plt.title("Books Average Rating vs Book Rating Count")
```

---
**Which authors generate the highest total gross sales?**

```python
total_gross_sales_by_author= df.groupby("Author")["gross sales"].sum()
total_gross_sales_by_author.sort_values(ascending=False).head(20).plot(kind="bar")
plt.xlabel("Author")
plt.ylabel("Total Gross Sales")
plt.title("Total Gross Sales for Each Author")
```

---
**How do units sold vary across different author ratings?**

```python
sns.boxplot(x="Author_Rating", y="units sold", data=df)
plt.xlabel("Author Rating")
plt.ylabel("Units Sold")
plt.title("Box Plot of Units Sold for each Author Rating")
```

---
**How have total units sold changed over the publishing years?**

```python
df.groupby("Publishing Year")["units sold"].sum().plot(kind="line", marker="o")
plt.xlabel("Publishing Year")
plt.ylabel("Total Units Sold")
plt.title("Total Units Sold over the Years")
```
